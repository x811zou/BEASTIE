```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
from run_model_stan_wrapper import parse_stan_output
from intersect_hets import summary_statistics
```

```python
def Visualization_theta(prefix,workDir,filters):
    outDir=workDir+prefix+"/output/"+filters
    outfilename=outDir+"/"+prefix+"_ASE_all.tsv"
    ASE_all=pd.read_csv(outfilename,sep="\t",header=0,index_col=False)
    # plot
    plt.hist(ASE_all['posterior_median'], density=True, bins=50)  # density=False would make counts
    plt.ylabel('Posterior Median '+r"(${\Theta}$)",size=15)
    plt.xlabel(filters,size=15)
    plt.axvline(x=1,color='r', linestyle='--')
    plt.title("Distribution of "+r"${\Theta}$"+" in "+prefix,size=20)
    # summary statistics
    summary_statistics(ASE_all['posterior_median'].tolist(), filters)
    return ASE_all
    
```
